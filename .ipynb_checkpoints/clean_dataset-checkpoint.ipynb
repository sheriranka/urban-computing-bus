## import libraries

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')


#sns.set_palette("husl")
#sns.color_palette("husl", 8)
```

## read data and store

```python
df = pd.read_csv("bus.csv")
df
```

## clean and fix data a bit

*Constants*

```python
#seed for random vals
seed = 42
#num of stops
p = 14
```

*Set stops*

```python
#estimate stops
km = KMeans(n_clusters=p, init='random', n_init=10, max_iter=20, random_state=seed)
loc = df[['lat','lon']]
km.fit(loc)
loc['label'] = km.labels_
sns.scatterplot(data=loc, x='lat', y='lon', hue='label',palette="tab20c")
sns.scatterplot(x=km.cluster_centers_[:, 0], y=km.cluster_centers_[:,1], color='red', marker='x')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.title("KMeans - Stop Estimation")
plt.show()
```

```python
#stop labels
#ADJUST THIS BASED ON KMEANS GRAPH
#LAST VALUE IS ALWAYS 15
p0 = [9,11,4,3,19,5,1,13,2,10,20,24,23,6,15]
#attach column and adjust names
print(loc["label"])
df['stop'] = loc['label'].apply( lambda d : p0[d])
```

```python
#drop incorrect data
df_rm = df[df.index < 26]
df = df.drop(df_rm.index, axis=0)
df = df.reset_index()
```

*Calculate times*

```python
#elapsed time
elapsed = [0]
for i in range(1, len(df["time"])):
    t = pd.to_datetime(df["time"][i],format='%H:%M:%S') - pd.to_datetime(df["time"][i-1],format='%H:%M:%S')
    elapsed.append(t.total_seconds())

df['elapsed'] = elapsed

#add time -> seconds column
ts = df['time'].apply( lambda d : pd.to_datetime(d,format='%H:%M:%S'))
ts = ts.apply( lambda d : d.time() )
sc = ts.apply( lambda d : (d.hour * 60 + d.minute) * 60 + d.second)
df['time_Seconds'] = sc

#df['Time_Seconds'] = df['Time'].total_seconds()
df
```

*More cleaning*

```python
#remove unnecessary columns
df = df.drop(columns=['id','acce','index'])
```

```python
print(df.to_string())
```

```python
#fix more row issues
for index, row in df.iterrows():
    #if first run of stop 2 isnt after stop 1, replace with 15
    #arbitrary number
    if row['stop'] == 2 & df.iloc[index-1]['stop'] != 1:
        row['stop'] = 15  

#add new empty column
df['round_trip'] = pd.NA

#calculate times between full round stop

for s in p0:
    df_r = df[df['stop'] == s]
    df['round_trip'] = df_r.time_Seconds.diff(df_r.time_Seconds.shift())


#set negative times to NaN
df['elapsed'] = df['elapsed'].mask(df['elapsed'] < 0)
df['round_trip'] = df['round_trip'].mask(df['round_trip'] < 0)

```

## observe data spread

```python
df_c = df[['elapsed', 'Time_Seconds']]
cor = df_c.corr()
print("Correlation between Time and Time Elapsed")
print(cor)

scaler = StandardScaler()
df_cs = scaler.fit_transform(df_c.values)

sns.kdeplot(x=df["elapsed"], color='blue', label="Time elapsed")
sns.kdeplot(x=df_cs[:,0], color='red', label="Time elapsed standarized")
plt.legend(loc="upper left")
plt.title("Time elapsed distribution")
plt.show()

sns.scatterplot(data=df, x='Parada', y='elapsed',hue="Parada",palette="tab10").set_xticks(df["Parada"])
plt.title("Time distribution for each stop from previous stop")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

## identify trends between time elapsed and time of day + find trends with stops

```python
#kmeans dist

km = KMeans(n_clusters=df["Parada"].unique().size, init='random', n_init=20, max_iter=200, random_state=seed)
km.fit(df_c)
df_c['label'] = km.labels_
sns.scatterplot(data=df_c, x='elapsed', y='Time_Seconds', hue='label',palette="tab10")
sns.scatterplot(x=km.cluster_centers_[:, 0], y=km.cluster_centers_[:,1], color='red', marker='x')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.title("KMeans")
plt.show()

df_c = df[['elapsed', 'Time_Seconds']]
#dbscan
#uses scaled data

gmm = GaussianMixture(n_components=df["Parada"].unique().size, max_iter=200)
gmm_cluster_labels = gmm.fit_predict(df_c)
df_c['label'] = gmm_cluster_labels
sns.scatterplot(data=df_c, x='elapsed', y='Time_Seconds', hue='label',palette="tab10")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.title("Gaussian Mix")
plt.show()

df_c = df[['elapsed', 'Time_Seconds']]

#actual distribution
sns.scatterplot(data=df, x='elapsed', y='Time_Seconds', hue='Parada',palette="tab10")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.title("Paradas prelabeled")
plt.show()
```
