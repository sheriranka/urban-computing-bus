```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('database_8_9_time_slot.csv')
```

```python
data
```

```python
time_slot = 9
start = 20
end = 13

row = data[data['time_slot'] == time_slot]

eta_no_bus = 0

route = [4, 5, 6, 13, 24, 23, 20, 19, 9, 10, 11, 2.1, 1, 2.2, 3]

eta_dict = {}

for index, r in row.iterrows():
    eta_no_bus += r['eta_bus']
    #print(r)
    eta_dict[r['stop_id']] = r
    
print('ETA one round', eta_no_bus/60)
```

```python
eta_dict
```

```python
ind = route.index(start)

st = route[ind]
count = 0

eta_s_e = 0

while st != end:
    ind += 1
    count += 1
    if(ind == len(route)):
        ind = 0
    st = route[ind]
    
    eta_s_e += eta_dict[float(st)]['eta_bus']
    
    
    
print(st, count)

print(eta_no_bus/60)

print(eta_s_e/60)
```
